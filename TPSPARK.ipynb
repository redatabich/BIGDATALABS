```python
from pyspark import SparkContext
from pyspark.streaming import StreamingContext

# Cr√©ation du SparkContext
sc = SparkContext(master="local[2]", appName="WordCountDStream")

# Batch interval : 5 secondes
ssc = StreamingContext(sc, 5)

# Source : socket
lines = ssc.socketTextStream("localhost", 9999)

# WordCount
words = lines.flatMap(lambda line: line.split(" "))
pairs = words.map(lambda word: (word, 1))
counts = pairs.reduceByKey(lambda a, b: a + b)

# Affichage
counts.pprint()

# Lancement
ssc.start()
ssc.awaitTermination()
```

```python
############################################
# TP : Spark Streaming avec DStream
# Environnement : Google Colab / PySpark
############################################

# ==============================
# 1. Installation de PySpark
# ==============================
#!pip install pyspark


# ==============================
# 2. Cr√©ation d‚Äôun serveur socket
#    simulant un flux de donn√©es
# ==============================
import socket
import time
import threading
import random

def start_socket_server():
    host = "localhost"
    port = 9999
    s = socket.socket()
    s.bind((host, port))
    s.listen(1)
    conn, addr = s.accept()

    # Messages simulant un flux temps r√©el
    messages = [
        "spark streaming dstream",
        "spark spark streaming",
        "big data spark"
    ]

    # Envoi al√©atoire des messages
    while True:
        msg = random.choice(messages)
        conn.send((msg + "\n").encode())
        time.sleep(2)

# Lancement du serveur en arri√®re-plan
threading.Thread(target=start_socket_server, daemon=True).start()


# ==============================
# 3. Cr√©ation du StreamingContext
# ==============================
from pyspark import SparkContext
from pyspark.streaming import StreamingContext

# Initialisation de Spark
sc = SparkContext.getOrCreate()

# Batch interval = 5 secondes
ssc = StreamingContext(sc, 5)


# ==============================
# 4. Lecture des donn√©es (DStream)
# ==============================
lines = ssc.socketTextStream("localhost", 9999)


# ==============================
# 5. WordCount en streaming
# ==============================
# D√©coupage des lignes en mots
words = lines.flatMap(lambda line: line.split(" "))

# Transformation en (mot, 1)
pairs = words.map(lambda w: (w, 1))

# Comptage des occurrences
counts = pairs.reduceByKey(lambda a, b: a + b)


# ==============================
# 6. Affichage des r√©sultats
# ==============================
counts.pprint()


# ==============================
# 7. D√©marrage et arr√™t du streaming
# ==============================
ssc.start()
ssc.awaitTerminationOrTimeout(30)
ssc.stop(stopSparkContext=False)


############################################
# QUESTIONS ‚Äì R√âPONSES
############################################

# Q4. Qu‚Äôest-ce qu‚Äôun micro-batch dans DStream ?
# Un micro-batch est un petit lot de donn√©es collect√©
# pendant un intervalle de temps fixe (ici 5 secondes)
# et trait√© comme un RDD.

# Q5. Sur quelle structure repose un DStream ?
# Un DStream repose sur une suite de RDD.
# Chaque micro-batch correspond √† un RDD.

# Q6. Quelle est la dur√©e du batch utilis√©e ?
# La dur√©e du batch est de 5 secondes.
```
